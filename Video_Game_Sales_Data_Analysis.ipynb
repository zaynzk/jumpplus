```python
import pandas as pd
import seaborn as sns


import matplotlib.pyplot as plt
```

```python
fileOne = 'Video_Games_Sales_as_at_22_Dec_2016.csv'
fileTwo = 'Video_Games_Sales_Truncated (1).csv'
df1 = pd.read_csv(fileOne)
df2 = pd.read_csv(fileTwo)
```

```python
# Show first five records
print("First five records of df1:")
print(df1.head())

print("\nFirst five records of df2:")
print(df2.head())

# Show all columns of the dataframe
print("Columns of df1:")
print(df1.columns)

print("\nColumns of df2:")
print(df2.columns)

# Show last five rows
print("Last five rows of df1:")
print(df1.tail())

print("\nLast five rows of df2:")
print(df2.tail())
```

```python
# Concatenate the dataframes
concatenated_df = pd.concat([df1, df2])

# Create the plot
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))  # Increase the plot size
platform_counts = concatenated_df['Platform'].value_counts()
platform_counts[:20].plot(kind='bar')
plt.title('Top 20 Platform Dominance')
plt.xlabel('Platform')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)
sns.despine()
plt.show()
```

```python
# Insight 2: Genre Preferences
plt.figure(figsize=(10, 6))
sns.countplot(data=pd.concat([df1, df2]), x='Genre')
plt.title('Genre Preferences')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Create the plot
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))  # Increase the plot size
sns.barplot(data=concatenated_df, x='Genre', y='Global_Sales')
plt.title('Global Sales Variation by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.xticks(rotation=45)
sns.despine()
plt.show()





```

```python
# Insight 4: Incomplete Data
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))  # Increase the plot size
missing_data_counts = concatenated_df.isnull().sum()
missing_data_counts.plot(kind='bar')
plt.title('Missing Data Counts')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=45)
sns.despine()
plt.show()
```

```python
# Concatenate the dataframes and reset index
concatenated_df = pd.concat([df1, df2])
concatenated_df.reset_index(drop=True, inplace=True)

# Insight 5: Global Sales Over Time
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=concatenated_df, x='Year_of_Release', y='Global_Sales')
plt.title('Global Sales Over Time')
plt.xticks(rotation=45)
sns.despine()
sns.reset_orig()
plt.show()
```
